```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read all data 
A = pd.read_html('Report16Jul2021121540.html')
#A
```

```python
# choosing the data that we will use 
A1=A[11]
#A1
```

```python
# changing the first row to be a header for all data
new_header1 = A1.iloc[0] #grab the first row for the header
A2 = A1[1:] #take the data less the header row
A2.columns = new_header1 #set the header row as the df header
A2 = A2.reset_index(drop=True)#reseting index
A2
```

```python
Indicator = A2.copy()#creating a copy
Indicator
```

```python
Indicator['Date'] = pd.to_datetime(Indicator.Date)#changing date format
Indicator
```

```python
#changing datatype to float
Indicator['High']=Indicator.High.astype('float')
Indicator['Close']=Indicator.Close.astype('float')
Indicator['Low']=Indicator.Low.astype('float')
Indicator['Open']=Indicator.Open.astype('float')
```

```python
#deleting columns not needed
del Indicator['Volume'],Indicator['TotalDistributionYield'], Indicator['InterestYield'], Indicator['CapitalPaymentYield'], Indicator['EY'], Indicator['DY'], Indicator['P/E']
#Indicator
```

```python
Indicator.describe()#descriptive statistics
```

```python
Indicator['Direct'] = (Indicator['Close'] - Indicator['Close'].shift(1)) #generating target variable
Indicator
```

```python
sig = []
for h in range (len(Indicator)):
    if Indicator['Direct'][h] > 0:#sell
        sig.append(3)
    else:
        sig.append(1)#buy
    
Indicator['signal'] = sig
Indicator
```

```python
def RSI(value):
    '''
    The function calculates the relative strength index parameters 
    to be used to calculate 14_day rsi values
    '''
    up = value[value>0].mean()
    down = -value[value<0].mean()
    return (100 * up / (up + down))

Indicator['Direction'] = (Indicator['Close'] - Indicator['Close'].shift(1)) #adding momentum
Indicator['RSI_14D'] = Indicator['Direction'].rolling(center=False, window=14).apply(RSI) #14D RSI
```

```python
#Indicator['RSI_14D']
```

```python
l = []
for i in range (len(Indicator)):
    if Indicator['RSI_14D'][i] >= 28 and Indicator['RSI_14D'][i] <= 30: # values below the signal line of 30 oversold
        l.append(1) #buy
    elif Indicator['RSI_14D'][i] >= 31 and Indicator['RSI_14D'][i] <= 69:
        l.append(2) #hold
    elif Indicator['RSI_14D'][i] >= 70:
        l.append(3)#sell
    else:
        l.append(2) #hold signal
Indicator['RSI_sig'] = l
```

```python
sns.countplot(x = 'RSI_sig', data=pd.DataFrame(Indicator), hue='RSI_sig')
#plt.savefig("aspen_rsicount")
plt.show()
Indicator['RSI_sig'].value_counts() #counts according to signal code 
```

```python
r = Indicator['RSI_14D']
h = [0,1300]
b = [30,30]
f = plt.plot(r, label = 'RSI_14D')
e = [70,70]
g = plt.plot(h,b,h,e) + f
plt.ylabel('Threshold value')
plt.xlabel('Days')
plt.legend()
#plt.savefig("aspen_rsi")
plt.show(g)
```

```python
n=21
Indicator['STOK'] = ((Indicator['Close'] - Indicator['Low'].rolling(window=n,center=False).mean()) /
                     (Indicator['High'].rolling(window=n, center=False).max() - 
                      Indicator['Low'].rolling(window=n, center=False).min())) * 100 #D% stochastic
Indicator['STOD'] = Indicator['STOK'].rolling(window = 7, center=False).mean()  #K% stochastic 
```

```python
m = []
for i in range (len(Indicator)):
    if Indicator['STOK'][i] >= 28 and Indicator['STOK'][i]<=30:
        m.append(1) #buy
    elif Indicator['STOK'][i] > 30 and Indicator['STOK'][i] <= 69:
        m.append(2) #sell    
    elif Indicator['STOK'][i] >= 70:
        m.append(3) #sell
    else:
        m.append(2) #hold
Indicator['STO_sig'] = m
```

```python
sns.countplot(x = 'STO_sig', data=pd.DataFrame(Indicator), hue='STO_sig')
#plt.savefig("aspen_stocount")
plt.show()
Indicator['STO_sig'].value_counts()
```

```python
t = Indicator['STOD']
j = Indicator['STOK']
u = plt.plot(t, label = 'STOD')
s = plt.plot(j, label = 'STOK')
h = [0,1300]
b = [30,30]
e = [70,70]
g = plt.plot(h,e,h,b) + u + s
plt.ylabel('Threshold value')
plt.xlabel('Days')
plt.legend()
#plt.savefig("aspen_sto")
plt.show()
```

```python
Indicator['EMA_10'] = Indicator['Close'].ewm(span=10,adjust=False,min_periods=1).mean() #50day exponential ma
Indicator['EMA_30'] = Indicator['Close'].ewm(span=20,adjust=False,min_periods=1).mean() #100day exponential ma
Indicator['EMA_90'] = Indicator['Close'].ewm(span=90,adjust=False,min_periods=1).mean() #200day exponential ma
```

```python
# append 1 for a buy, 0 for a sell and 2 for hold and NaN
n = []
for i in range (len(Indicator)):
    if Indicator['EMA_10'][i] < Indicator['EMA_30'][i] and Indicator['EMA_90'][i]:
        n.append(1) #buy
    elif Indicator['EMA_10'][i] > Indicator['EMA_30'][i] and Indicator['EMA_90'][i]:
        n.append(3) #sell
    elif Indicator['EMA_10'][i] == Indicator['EMA_30'][i] == Indicator['EMA_90'][i]:
        n.append(2)  #hold      
    else:
        n.append(2)#hold
Indicator['EMA_sig'] = n     
```

```python
sns.countplot(x = 'EMA_sig', data=pd.DataFrame(Indicator), hue='EMA_sig')
#plt.savefig("aspen_emacount")
plt.show()
Indicator['EMA_sig'].value_counts()
```

```python
# buy if 50EMA line cuts the other two EMA lines from above and moves below both the EMAs
# sell if 50EMA line cuts the other two EMA lines from below and moves above both the EMAs
a = Indicator['EMA_10'] #50 day exponential moving average
b = Indicator['EMA_30'] #100 day exponential moving average
c = Indicator['EMA_90'] #200 day exponential moving average
d = plt.plot(a, label = 'EMA_10' ) + plt.plot(b, label = 'EMA_30') + plt.plot(c, label = 'EMA_90')
plt.ylabel('Threshold value')
plt.xlabel('Days')
plt.legend()
#plt.savefig("aspen_ema")
plt.show(d)
```

```python
Indicator['26_ema'] = Indicator['Close'].ewm(span=26,adjust=True,ignore_na=False).mean() #26day ema
Indicator['12_ema'] = Indicator['Close'].ewm(span=12,adjust=True,ignore_na=False).mean() #12day ema
Indicator['MACD'] = Indicator['12_ema'] - Indicator['26_ema'] #macd
Indicator['9EMA_MACD'] = Indicator['MACD'].ewm(span=9,adjust=True,ignore_na=False).mean() #9daymacd_ema
#Indicator
```

```python
f = []
for i in range (len(Indicator)):
    if Indicator['MACD'][i] > Indicator['9EMA_MACD'][i]:
        f.append(1) #buy
    elif Indicator['MACD'][i] < Indicator['9EMA_MACD'][i]:
        f.append(3) #sell
    elif Indicator['MACD'][i] == Indicator['9EMA_MACD'][i]:
        f.append(2) #
    else:
        f.append(2)
Indicator['MACD_sig'] = f 
```

```python
sns.countplot(x = 'MACD_sig', data=pd.DataFrame(Indicator), hue='MACD_sig')
#plt.savefig("aspen_macdcount")
plt.show()
Indicator['MACD_sig'].value_counts()
```

```python
#imputing values
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
Database = my_imputer.fit_transform(Database)
```

```python
u = pd.DataFrame(Database, columns = ['EMA_10', 'RSI_14D', 'MACD', 'STOK', 'signal'])
u[513:1109]
```

```python
trial_test=u[364:513]#test set
X_test = trial_test[['EMA_10', 'RSI_14D', 'MACD', 'STOK']]
Y_test = trial_test[['signal']]
#first = u[0:364]
trial_train = u[513:1109]#trainset
X_train = trial_train[['EMA_10', 'RSI_14D', 'MACD', 'STOK']]
Y_train = trial_train[['signal']] 
#trial_train=pd.concat([first, second], axis = 0)
#u.loc[row_indexer,column_indexer]
```

```python
trial_test['signal'].value_counts()
```

```python
from sklearn.metrics import precision_score,accuracy_score,roc_curve, f1_score,log_loss
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import model_selection
```

```python
classifyers = [
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    BaggingClassifier(),
    KNeighborsClassifier(),
    MLPClassifier()
    
]
models = []
for classifyer in classifyers:
    classifyer.fit(X_train,Y_train)
    print(f"{classifyer.__class__.__name__:10}: trained")
    models.append(classifyer)
```

```python
models
```

```python
#classification report
from sklearn.metrics import classification_report
for model in models:
    Y_pred = model.predict(X_test)
    print(
    f"{model.__class__.__name__:22}: classification_report = {classification_report(Y_test, Y_pred)} \n"
    )
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
for model in models:
    Y_pred = model.predict(X_test)
    print(
    f"{model.__class__.__name__:22}: confusion_matrix(= {confusion_matrix(Y_test, Y_pred)} \n"
    )
```

```python
#roc_auc score
from sklearn.metrics import roc_auc_score
for model in models:
    Y_pred = model.predict(X_test)
    print(
    f"{model.__class__.__name__:22}: roc_auc_score(= {roc_auc_score(Y_test, Y_pred)} \n"
    )
```
